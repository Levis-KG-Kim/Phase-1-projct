3. DATA PREPARATION | CLEANING UP THE DATA

DROP ROWS AND COLUMNS

```python
# Getting the sum of null values per column
df.isna().sum()
```

```python
# Getting the relevant columns
df = df[['Investigation.Type',  'Event.Date',
       'Location', 'Country','Airport.Name', 'Aircraft.damage',
       'Aircraft.Category', 'Make', 'Model','Amateur.Built',
        'Number.of.Engines', 'Engine.Type', 'Weather.Condition','Total.Fatal.Injuries',
       'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured',
        'Report.Status']]
```

```python
# Get only the rows with Aircraft category as "Airplane" 
df = df[df["Aircraft.Category"] == "Airplane"]
```

```python
# Get data from 2000 to-date
df = df.loc[(df['Event.Date'] >= '2000-01-01')
                     & (df['Event.Date'] < '2024-09-15')]
```

```python
# get rows where pilot was not involved
df = df.loc[(df["Report.Status"] == "Foreign") | (df["Report.Status"] == "Probable Cause")]
```

```python
# Drop rows by column value 
df = df[df["Investigation.Type"] == "Accident"]
df = df[df["Engine.Type"] != "Unknown"]
df = df[df["Make"] != "Unknown"]
```

```python
# Check and drop duplicates
df.duplicated().sum()
```

ADDING COLUMNS

```python
# Split the date to get year as a new column
df["Event.Year"]=df["Event.Date"].dt.year
```

```python
# Add all injured to get total injured as a new column
df["Total.Injured"] = df[["Total.Fatal.Injuries", "Total.Serious.Injuries", "Total.Minor.Injuries"]].sum(axis=1)
```

FINAL CLEANING

```python
# Rename columns
df.rename(columns={"Aircraft.damage": "Aircraft.Damage"}, inplace = True)
```

```python
# Reorder columns
df = df[['Event.Date','Event.Year','Airport.Name', 'Location', 'Country', 'Aircraft.Damage', 'Make', 'Model',
       'Amateur.Built', 'Number.of.Engines', 'Engine.Type',
       'Weather.Condition','Total.Injured','Total.Fatal.Injuries', 'Total.Serious.Injuries',
       'Total.Minor.Injuries', 'Total.Uninjured', 'Report.Status']]
```

```python
# Reset index
df.reset_index(drop=True)
```
